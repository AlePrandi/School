```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./smartphones.csv")
data.head(10)
```

```python
apple = data[data["phone_brand"] == "apple"]
samsung = data[data["phone_brand"] == "samsung"]
price_apple = apple["price_usd"]
price_samsung = samsung["price_usd"]

fig, ax = plt.subplots(figsize=(10,5))

ax.hist(price_samsung, bins=50, color="blue", edgecolor="k",label = "price samsung", alpha = 0.7)
ax.hist(price_apple, bins=50, color="red",edgecolor="k", label = "price apple",alpha = 0.7)
ax.set_xlabel("price_usd")
ax.set_ylabel("frequency")

ax.legend()
plt.plot()
```

```python
# vendite negli anni di apple e samsung


apple_sales = data[data["phone_brand"] == "apple"].groupby("anno").size()
samsung_sales = data[data["phone_brand"] == "samsung"].groupby("anno").size()

fig, ax = plt.subplots(figsize=(7,4))
ax.plot(apple_sales, label = "Apple", color = "red")
ax.plot(samsung_sales, label = "Samsung", color = "blue")

ax.set_xlim(2018)
ax.set_title("Andamento vendite negli anni")
ax.set_xlabel("anno")
ax.set_ylabel("vendite")
ax.legend()
plt.show()
```
