```python
NomeComune = "busca".upper()
comune = residenza[residenza["ComuneRES"] == NomeComune][["AnnoS","SESSO","Lau"]]
ComuneF = comune[comune["SESSO"] == "F"][["AnnoS","Lau"]]
ComuneM = comune[comune["SESSO"] == "M"][["AnnoS","Lau"]]

fig, ax = plt.subplots(figsize= (10,4))

ax.plot(ComuneF["AnnoS"], ComuneF["Lau"], label = "Femminile", color = "violet", linewidth = 1.5)
ax.plot(ComuneM["AnnoS"], ComuneM["Lau"], label = "Maschile", color = "lightblue", linewidth = 1.5)

ax.set_title(f"Laureati a {NomeComune} per genere dal 2010 al 2023")
ax.legend()
fig.set_facecolor("lightblue")
ax.grid(True)
plt.show()
```
