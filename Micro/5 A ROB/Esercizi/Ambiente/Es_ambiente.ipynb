```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataComuni = pd.read_csv("regpie-Asco_siti_contaminati_6051-all.csv", sep=";")
dataComuni
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.hist(dataComuni["prov"], edgecolor="black", bins=len(dataComuni["prov"].unique()))
ax.set_xticks(ax.get_xticks(), ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
attivi = dataComuni[dataComuni["stato_procedimento"] == "ATTIVO"]["prov"]
conclusi = dataComuni[dataComuni["stato_procedimento"] == "CONCLUSO"]["prov"]

fig, ax = plt.subplots(figsize=(10,4))

ax.hist(attivi, edgecolor="black", bins = len(dataComuni["prov"].unique()), color = "red", alpha = 0.5)
ax.hist(conclusi, edgecolor="black", bins = len(dataComuni["prov"].unique()), color = "green", alpha = 0.5)
ax.set_xticks(ax.get_xticks(), ax.get_xticklabels(), rotation=45, ha='right')

ax.set_title("siti contaminati attivi e bonificati per provincia")
ax.legend(["attivi", "bonificati"])
ax.set_ylabel("numero di siti")
fig.set_facecolor("lightgrey")
plt.show()
```
