## Grafici con `matplotlib.pyplot`

un grafico ha degli elementi principali:
- la figura in cui si possono rappresentare pi√π grafici
- gli assi con le loro etichette
- la griglia
- il titolo

link: https://matplotlib.org

esempi: https://matplotlib.org/stable/gallery/index

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots() # √® il metodo pi√π potente rispetto a plt.plot
x = np.linspace(-5,5,1000)
y = np.exp(-x**2) 
x_1 = np.linspace(-7,7,1000)
y_1 = np.exp(-x**2) 
# ax.set_facecolor("orange") colore dello sfondo
ax.plot(x,y, linewidth=2, color= np.array([0.5,0.1,0.4]))
ax.plot(x_1,y_1, linewidth=2, linestyle="--" , color= np.array([0.5,0.1,0.4])) 
#si pu√≤ usare il colore come stringa, RGB, RGBA o codice esadecimale
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("il Grafico di una curva Gaussiana")
ax.grid()
plt.show()
```

```python
n_punti = 100
x = np.random.uniform(0.,1., n_punti)
y = np.random.uniform(0.,1., n_punti)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y,linestyle=" ", marker="o",color="black", markersize= 5)
plt.show()
```

# Grafici a dispersione

```python
n_dots = 100
x = np.random.uniform(0.,1.,n_dots)
y = np.random.uniform(0.,1.,n_dots)
size = np.random.uniform(2, 20, n_dots)
color = np.random.choice(["white","yellow"], n_dots)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
#fig.set_facecolor("#FEFCFB")
fig.set_facecolor("#1282A2")
ax.set_facecolor("#0A1128")
ax.scatter(x,y,c = color, s = size, marker= "*", alpha= 0.5) # plot ha il limite che tutti i punti devono essere graficamente uguali
plt.show()
```

```python
punti = 400
x, y = np.random.uniform(0.,1., punti),np.random.uniform(0.,1., punti)
x_1, y_1 = np.random.uniform(0.,1., punti),np.random.uniform(0.,1., punti)
x_2, y_2 = np.random.uniform(0.,1., punti),np.random.uniform(0.,1., punti)

size = np.random.uniform(100, 200, punti)

fig, ax = plt.subplots(figsize=(5,5))

ax.scatter(x,y,c = "blue", s = size, alpha= 0.4, label = "tab:blue", linewidths=0)
ax.scatter(x_1,y_1,c = "green", s = size, alpha= 0.4, label = "tab:green", linewidths=0)
ax.scatter(x_2,y_2,c = "orange", s = size, alpha= 0.4, label = "tab:orange", linewidths=0)
ax.legend(loc="upper right")
ax.grid()
plt.show()
```

# Istogrammi

```python
n_lanci = 10000
dado1 = np.random.randint(1,7,n_lanci)
dado2 = np.random.randint(1,7,n_lanci)
dado3 = np.random.randint(1,7,n_lanci)
dado4 = np.random.randint(1,7,n_lanci)
  
fig, ax = plt.subplots(figsize=(5,5))
#fig.set_facecolor("#FEFCFB")
fig.set_facecolor("#1282A2")
ax.set_facecolor("#0A1128")
ax.hist(dado1+dado2+dado3+dado4, bins=len(np.unique(dado1+dado2+dado3+dado4)), color="white",edgecolor="black")
plt.show()
```
