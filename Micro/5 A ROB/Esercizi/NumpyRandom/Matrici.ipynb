# Esercizio

Creare la cornice nera dell'immagine

```python
cornice = np.zeros((dim,dim), dtype=np.float64)

#bordo esterno
cornice[0] = 1
cornice[-1] = 1
cornice[:, 0] = 1
cornice[:, -1] = 1

#rettangolo interno
#cornice[3:9, 7:16] = 1
```

```python
plt.imshow(cornice, cmap="Greys")
```

## Esercizio 

Disegnare un bersaglio di quadrati concentrici (si pu√≤ usare al massimo un ciclo FOR)

Bonus: i cerchi devono avere colori diversi

```python
cornice = np.zeros((dim,dim), dtype=np.float64)
for n in range(0, dim):
    cornice[n: dim-n, n: dim - n ] = 0
    if n % 2 == 0:
        cornice[n: dim-n, n: dim - n ] = 1
```

```python
plt.imshow(cornice, cmap="Greys")
```
