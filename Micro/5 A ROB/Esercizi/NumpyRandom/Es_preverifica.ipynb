# Funzione aggregazione

```python
n_alunni = 27
n_voti = 10
voti = np.random.randint(2,11,(n_alunni,n_voti))
media_alunni = np.mean(voti,axis=1) > 6
print(f"Medie sopra al 6: {np.count_nonzero(media_alunni)}")
```

```python
n_classi = 1000
alunni = 25
mat_classi = np.random.randint(1,366,(n_classi,alunni), dtype=np.int16)
np.sum([len(np.unique(classe)) < alunni for classe in mat_classi]) # np.unique restituisce i valori unici dell'array
```
