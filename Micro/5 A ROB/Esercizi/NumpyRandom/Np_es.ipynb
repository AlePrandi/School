```python
x = np.linspace(-5,5,400) #400 valori equispaziati tra -5 e 5
y = np.exp(x)
plt.plot(y)
```

```python
x = np.linspace(0, 4*np.pi, 400)
y = np.sin(x)
plt.plot(y)
```

```python
y[y > 0.5] = 0.5
plt.plot(x,y)
```

```python
x = np.linspace(-20*np.pi, 20*np.pi, 1000)
y = np.sin(x)/x
plt.plot(x,y)
```

```python
x = np.linspace(-5,5, 1000)
y = np.exp(-x**2) #distribuzione gaussiana
plt.plot(x,y) 
```

## Esercizio


### Generazione numeri pseudo casuali usando la distribuzione gaussiana usando `np.random.normal()`

```python
n_persone = 20
media = 175
altezze_cm = np.random.normal(media, 8, n_persone) # random.normal genera numeri usando una distribuzione gaussiana
np.round(altezze_cm, 1)
```

### Estraiamo in maniera pseudo-casuale elementi da un array con la funzione `np.random.choice()`

```python
Moneta = np.array(["Testa", "Croce"]) #array di stringhe
prob = np.array([0.7,0.3])
np.random.choice(Moneta, 10, p = prob)
```
