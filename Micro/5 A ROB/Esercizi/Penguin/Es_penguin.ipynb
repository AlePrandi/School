## Esercizio con dati veri

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# pandas permette di leggere i dati
```

```python
data = pd.read_csv("./penguins_size.csv")
data.head()
```

```python
data
```

```python
data.dropna(inplace=True)
# .values lo fa diventare un array di Numpy
body_mass_Adelie = data[data["species"] == "Adelie"]["body_mass_g"].values
body_mass_Adelie
```

```python
body_mass_Gentoo = data[data["species"] == "Gentoo"]["body_mass_g"].values
body_mass_Gentoo
```

```python
bins = np.linspace(2500, 6500, 20)
fig, ax = plt.subplots(figsize=(4, 4))

ax.hist(body_mass_Adelie, label="Adelie", edgecolor="black",
        alpha=0.8, bins=bins, color="#DD6E42")
ax.hist(body_mass_Gentoo, label="Gentoo", edgecolor="black",
        alpha=0.8, bins=bins, color="#4F6D7A")

ax.set_xlabel("Peso corporeo in (g)")
ax.set_ylabel("N° di esemplari")

ax.set_facecolor("#E8DAB2")
ax.legend()
# salva il grafico come immagine, pdf o altri tipi di file
plt.savefig("./IstoPenguin.png", dpi=200)
plt.show()
```

## Più grafici nella stessa figura

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4)) #axs al posto di ax quando uso più grafici
axs[0].hist(body_mass_Adelie, label="Adelie", edgecolor="black",
        alpha=0.8, bins=bins, color="#DD6E42")
axs[1].hist(body_mass_Gentoo, label="Gentoo", edgecolor="black",
        alpha=0.8, bins=bins, color="#4F6D7A")

for i in range(len(axs)):
    axs[i].set_xlabel("Peso corporeo in (g)")
    axs[i].set_ylabel("N° di esemplari")
    axs[i].set_facecolor("#E8DAB2")
    axs[i].legend()

plt.show()
```

```python
data.dropna(inplace=True)
data.head()
```

```python
data.info()
```

```python
cols = ["culmen_length_mm","culmen_depth_mm","flipper_length_mm", "body_mass_g"]

data[cols].head()
```

```python
spc = data["species"].values
spc
peng_colors = ["green", "blue", "red"]
spc_color = spc.copy()
spc_color[spc_color == "Adelie"] = peng_colors[0]
spc_color[spc_color == "Gentoo"] = peng_colors[1]
spc_color[spc_color == "Chinstrap"] = peng_colors[2]
```

```python
fig, axs = plt.subplots(4,4, figsize=(15,15))

for i, icol in enumerate(cols):
    for j, jcol in enumerate(cols):
        if i!= j:
            x = data[icol].values
            y = data[jcol].values
            axs[i, j].scatter(x, y, s = 5, c = spc_color)
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel(jcol)
            axs[i, j].set_facecolor("#E1D89F")
        else:
            x = data[icol].values
            axs[i, j].hist(x, color = "#6153CC", edgecolor = "black")
            axs[i, j].set_xlabel(icol)
            axs[i, j].set_ylabel(jcol)
            axs[i, j].set_facecolor("#E1D89F")
            
plt.show()
```

```python
marziani = pd.read_csv("./marziani.csv")
marziani.dropna(inplace=True)
marziani.head()
cols_marz = ["n_arti", "peso", "altezza", "larghezza"]
marziani[cols_marz].head()
```

```python
dati_meteo = pd.read_csv("./MonthElb_012010_004028900.csv", header= None, na_values=9999.9)
dati_meteo.head()
```

```python
cols_meteo = [5,8,9,10,13]
fig, axs = plt.subplots(5,5, figsize=(15,15))

for i, icol in enumerate(cols_meteo):
    for j, jcol in enumerate(cols_meteo):
        if i!= j:
            x = dati_meteo[icol].values
            y = dati_meteo[jcol].values
            axs[i, j].scatter(x, y, s = 3)
            axs[i, j].set_facecolor("#E1D89F")
        else:
            x = dati_meteo[icol].values
            axs[i, j].hist(x, edgecolor = "black")
            axs[i, j].set_facecolor("#E1D89F")
            
fig.suptitle("Analisi grafica meteorologica", fontsize = 30, y = 0.93)
plt.show()
```
