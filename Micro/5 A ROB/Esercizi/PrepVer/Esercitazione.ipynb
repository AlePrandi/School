## ESERCIZIO 1
Il file *10000Pi.txt* contiene le prime 10000 cifre del $\pi$ in base 10.

Crea un programma in Python 3 (decidi se usare Numpy oppure no) che legga il file e permetta di rispondere alle domande seguenti:
* nelle cifre del $\pi$ quante volte capita una sequenza 111?
* [FACOLTATIVO DSA] è presente la seguenza 999999? In caso di risposta affermativa individua in quale punto della sequenza di cifre decimali è presente.

```python
conta111 = 0
pos = 0
with open("./10000Pi.txt", "r") as file:
    f = file.read()
    sequenza = "111"
    for i in range(len(f) - len(sequenza) + 1):
        if f[i:i+len(sequenza)] == sequenza:
            conta111 += 1
    for i in range(len(f) - len("999999") + 1):
        if f[i:i+len("999999")] == "999999": 
            pos = i
            print(f"Sequenza 999999 trovata in pos: {pos}")
```

## ESERCIZIO 2
Rappresenta le seguenti funzioni $f(x)$ e $g(x)$ in un grafico che sia completo di tutti gli elementi necessari.

$ f(x) =   \left\{
\begin{array}{ll}
      e^x & x\le 0 \\
      \cos(2\pi x) & x \gt 0 \\
\end{array} 
\right.  $

$g(x) = \frac{1}{1+x^2}$

```python
x = np.linspace(-100, 100, 400)

f_x = np.where(x<= 0, np.exp(x), np.cos(2 * np.pi * x))
g_x = 1 / (1 + x**2)

plt.figure(figsize=(10, 6))
plt.plot(x, f_x, label=r'f_x')
plt.plot(x, g_x, label=r'g_x')
plt.legend()
plt.show()
```

## ESERCIZIO 3
Dato un numero intero n crea un array numpy unidimensionale fatto in questo modo:

    [1, -1, 2, -2, 3, -3..., n, -n]
    
Per esempio se n=3 l'array è:

    [1, -1, 2, -2, 3, -3]
    
**NON PUOI USARE CICLI, IF E LISTE**

```python
n = 3
array_pos = np.arange(1, n + 1) 
array_neg = -array_pos

array_final = np.zeros(n * 2)
array_final[0::2] = array_pos
array_final[1::2] = array_neg
array_final
```

### ESERCIZIO 6
Genera un array di Numpy contenente tutti i numeri di Fibonacci minori di 1000. Per generare l'array puoi usare un solo ciclo `for` oppure `while`.
I numeri della sequenza di Fibonacci sono ottenuti a partire dai primi due valori: *1, 1*. Ogni valore successivo a partire dal terzo è ottenuto sommando i due precedenti.

Usando l'array dei numeri di Fibonacci e facendo uso di opportune funzioni di Numpy calcola:
- l'array dei rapporti tra un numero di Fibonacci e il suo numero di Fibonacci precedente,
- l'array contenente solo i numeri di Fibonacci pari,
- [FACOLTATIVO DSA] l'array ottenuto calcolando il quadrato di ogni numero di Fibonacci meno il prodotto dei due numeri di Fibonacci adiacenti (precedente x successivo).

**NON PUOI USARE CICLI, IF E LISTE**

```python
n1,n2 = 1,1
array_fibonacci = 0
while n1 < 1000:
    array_fibonacci = np.append(array_fibonacci, n1)
    n1, n2 = n2, n1 + n2
    
array_fibonacci
```

```python
rapporti = array_fibonacci[1:] / array_fibonacci[:-1]
rapporti
```

```python
fibo_pari = array_fibonacci[array_fibonacci % 2 == 0]
fibo_pari
```

```python
array_quad = array_fibonacci[1:-1] ** 2 - array_fibonacci[:-2] * array_fibonacci[2:]
array_quad
```
