```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("popolazione.csv")
data
data["Population"] = data["Population"].str.replace(",", "").astype(float)
data["Year"] = data["Year"].astype(int)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(data["Year"], data["Population"])
plt.show()
```

```python
x = data["Year"].values.reshape(-1,1)
y = data["Population"].values.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)  

fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(x, y, color='blue', label="Dati reali", alpha=0.5)
ax.plot(x, y_pred, color='red', label="Regressione lineare", linewidth=2)
ax.set_xlabel("Anno")
ax.set_ylabel("Popolazione")
ax.set_title("Trend della Popolazione Mondiale")
ax.legend()
ax.grid(True)
plt.show()
```

```python
anno_futuro = np.array([[2030]]) #devo passare l'anno come una matrice
popolazione_2050 = model.predict(anno_futuro)
print(f"Popolazione stimata nel 2050: {popolazione_2050}")
```
