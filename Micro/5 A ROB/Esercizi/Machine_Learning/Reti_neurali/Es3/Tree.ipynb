```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('./registro_utilizzo_giacca.csv')
data.head()
```

```python
features = ["temperatura", "vel_vento"]
target = ["condizione"]

X = data[features].to_numpy()
y = data[target].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

**Per gli alberi decisionali NON serve normalizzare le feature**

```python
albero = tree.DecisionTreeClassifier(max_depth=10)
```

```python
albero.fit(X_train, y_train)
```

```python
albero.score(X_test, y_test)
```

```python
albero.predict([[22, 3]])[0]
```

```python
fig,ax = plt.subplots(figsize=(10,10))
tree.plot_tree(albero, feature_names=features, class_names=albero.classes_, filled=True, label="all", rounded=True)
plt.show()
```

Il più alto è il nodo **root**, i nodi senza figli si chiamano **foglie**, gli altri sono chiamati **nodi decisionali**, l'altezza dell'albero è data dal max_depth.

**GINI** è un valore che prende il nome dall'indice di gini che misura la purezza di un insieme, 

l'insieme è puro se contiene campioni di un solo tipo e in quel caso vale 0
