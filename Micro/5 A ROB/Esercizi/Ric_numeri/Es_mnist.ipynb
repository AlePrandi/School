## Esercizio mnist 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def string2face(s):
    Id = np.array([int(n) for n in s.split(" ")])
    return Id.reshape(24,24)
```

```python
dati = pd.read_csv("./mnist_784_csv.csv") # ogni riga ha un'immagine e ogni colonna Ã¨ un pixel
dati.head()

pixels_columns = dati.columns[:-1]
dati[pixels_columns]
digits = dati[pixels_columns].values
digits_names = dati["class"].values
```

## Esericizio

Sorteggiare 25 cifre a caso nel dataset e rappresentare graficamente in una figura, indicare nel titolo di ciascuna la cifra rappresentata

```python
cifre = np.random.choice(np.arange(0, len(digits)), 25 , replace=False)
cifra = digits[cifre]
cifra_num = digits_names[cifre]

fig, axs = plt.subplots(5,5,figsize=(15,15))
    
for i, ax in enumerate(axs.reshape(-1)):
    ax.imshow(cifra[i].reshape(28, 28), cmap="Greys")
    ax.set_title(cifra_num[i])
    ax.set_axis_off()

plt.show()
```

```python
diz_cifre = {}
for i in range(0, 10):
    diz_cifre[i] = digits[digits_names == i]

diz_cifre
```

```python
fig, axs = plt.subplots(10,10,figsize=(15,15))
    
for i in range(0,10):
    immagine_sel = diz_cifre[i]
    num = immagine_sel.shape[0]
    indici = np.random.choice(np.arange(0,num), 10, replace=False)
    immagine_sel = immagine_sel[indici]
    for col, img in enumerate(immagine_sel):
        ax = axs[i, col]  
        ax.imshow(img.reshape(28, 28), cmap="Greys") 
        ax.set_title(f"{i}") 
        ax.set_axis_off()

plt.show()
```
