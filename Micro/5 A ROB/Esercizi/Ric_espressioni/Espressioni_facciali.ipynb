## Espressioni facciali

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def string2face(s):
    FacciaId = np.array([int(n) for n in s.split(" ")])
    return FacciaId.reshape(48,48)
```

```python
dati = pd.read_csv("./fer2013.csv")
dati.head(10)
#pixels sono immagini di 48 x 48 pixels
diz_emozioni = { 0 : "Angry", 1 :"Disgust", 2 :"Fear", 3 : "Happy", 4 : "Sad", 5 : "Surprise", 6 : "Neutral"}

emotions = dati["emotion"].values
expression = dati["pixels"].values

id_faccia = 925

FacciaId = np.array([int(n) for n in expression[id_faccia].split(" ")])

fig, ax = plt.subplots(figsize=(3,3))
ax.imshow(FacciaId.reshape(48,48), cmap="Greys_r")
ax.set_title(diz_emozioni[emotions[id_faccia]])
ax.set_axis_off()
plt.show()
```

## Esercizio

Creare una figura con 25 espressioni facciali sorteggiate a caso. Per ognuna indicare l'emozione

```python
dati = pd.read_csv("./fer2013.csv")
dati.head(10)
#pixels sono immagini di 48 x 48 pixels
diz_emozioni = { 0 : "Angry", 1 :"Disgust", 2 :"Fear", 3 : "Happy", 4 : "Sad", 5 : "Surprise", 6 : "Neutral"}

emotions = dati["emotion"].values
expressions = dati["pixels"].values

ids = np.random.choice(np.arange(0, len(emotions)), 25 , replace=False)
chosen_emotions = emotions[ids]
chosen_expression = expressions[ids]

fig, axs = plt.subplots(5,5,figsize=(15,15))
    
for i, ax in enumerate(axs.reshape(-1)):
    ax.imshow(string2face(chosen_expression[i]), cmap="Greys_r")
    ax.set_title(diz_emozioni[chosen_emotions[i]])
    ax.set_axis_off()

plt.show()
```
