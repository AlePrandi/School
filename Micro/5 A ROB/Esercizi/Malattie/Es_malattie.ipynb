```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("heart_disease.csv")
data.head()
```

```python
fumatori = data[data["Smoking"] == "Yes"]
fumatori
```

```python
donneDiab = data[(data["Diabetes"] == "Yes") & (data["Gender"] == "Female")][["Blood_Pressure", "Age"]]
donneDiab
```

```python
maleFum = data[(data["Age"] > 18) & (data["Age"] < 35) & (data["Smoking"] == "Yes") & (data["Gender"] == "Male")]
maleFum
```

```python
fumatoriTot = data[data["Smoking"] == "Yes"].value_counts("Smoking")
fumatoriTot
```

```python
fumatoriGen = data[data["Smoking"] == "Yes"].value_counts("Gender")
fumatoriGen
```

```python
maFum = data[data["Heart_Disease_Status"] == "Yes"][["Gender", "Smoking", "Heart_Disease_Status"]].groupby(["Gender", "Smoking"]).count()
maFum
```

```python
oreSonno = data[["Sleep_Hours", "Gender"]].groupby("Gender").mean()
oreSonno
```

```python
fumatori_Maschi = data[(data["Gender"] == "Male") & (data["Smoking"] == "Yes")].copy()

intervalli = [18, 30, 45, 60, 80, 100]
etichette = ["18-30", "30-45", "45-60", "60-80", "80+"]

fumatori_Maschi["Age_Group"] = pd.cut(fumatori_Maschi["Age"], bins=intervalli, labels=etichette)

fig, ax = plt.subplots()
fumatori_Maschi["Age_Group"].value_counts().sort_index().plot(kind='bar', ax=ax, edgecolor="black")
ax.set_xlabel("Age Group")
ax.set_ylabel("Number of Smokers")
ax.set_title("Number of Male Smokers by Age Group")
ax.set_xticks(range(len(etichette)))
ax.set_xticks(ax.get_xticks(), ax.get_xticklabels(), rotation=45, ha='right')
ax.set_xticklabels(etichette)
plt.show()
```
