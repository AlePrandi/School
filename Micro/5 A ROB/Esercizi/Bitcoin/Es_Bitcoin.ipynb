```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dati_bitcoin = pd.read_csv("./BTC-Daily.csv")
dati_bitcoin.head()
```

```python
btc_high = dati_bitcoin["high"].values[::-1]
btc_low = dati_bitcoin["low"].values[::-1]
btc_close = dati_bitcoin["close"].values[::-1]
btc_volumeUSD = dati_bitcoin["Volume USD"].values[::-1]
tempo = dati_bitcoin["unix"].values[::-1]
```

## Es. creare una fig con due grafici:

- il 1° mostra l'andamento dei prezzi ai high e low giornalieri nel tempo
- il 2° mostra se c'è correlazione tra il prezzo di chiusura giornaliero e il volume

```python
fig, ax = plt.subplots(figsize=(10, 3))

ax.set_title("Andamento prezzi high e low")
ax.set_xlabel("Tempo")
ax.set_ylabel("Prezzo bitcoin")
ax.set_facecolor("black")

ax.plot(btc_high, label="High", color="green")
ax.plot(btc_low, label="Low", color="red", linewidth= 0.6, alpha = 0.6, linestyle = "--")

ax.legend()
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 3))
ax.scatter(btc_volumeUSD, btc_close, label="Close", color="blue", s = 2)

ax.set_xlabel("VolumeUSD")
ax.set_ylabel("prezzo Chiusura")
ax.legend()
plt.show()
```

```python
num_valori = 7200
movimenti = np.array([[0,1],[0,-1],[1,0],[-1,0]])
posizioni = np.cumsum(movimenti[np.random.randint(0,4, num_valori)], axis=0)

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(posizioni[:, 0], posizioni[:,1],s = 0.5, marker = "h")
ax.scatter(0,0, s = 30, c ="green")
ax.scatter(posizioni[-1,0], posizioni[-1,1], s = 30, c="red" )
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Cammino Aleatorio / Random walk")
ax.set_facecolor("pink")
fig.set_facecolor("yellow")
ax.text(0,0, "start", fontsize = 20)
ax.text(posizioni[-1,0], posizioni[-1,1], "finish", fontsize = 20)
plt.show()
```
