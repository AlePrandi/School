## Crea una cornice blu sull'immagine di spessore 10pixel


```python
image_cornice = image_crop.copy()
bordo = 10
image_cornice[:bordo,:, 2] = 255
image_cornice[:,:bordo, 2] = 255
image_cornice[-bordo:,:, 2] = 255
image_cornice[:, -bordo:, 2] = 255

image_cornice[:bordo,:, :2] = 0
image_cornice[:,:bordo, :2] = 0
image_cornice[-bordo:,:, :2] = 0
image_cornice[:, -bordo:, :2] = 0
plt.imshow(image_cornice)
```

### Es1

scala di grigi

```python
grigi = image.copy()
grigi[:,:, 0] = image[:,:,0] + image[:,:,1]+ image[:,:,2]
grigi[:, :, 1] = image[:,:,0] + image[:,:,1]+ image[:,:,2]
grigi[:, :, 2] = image[:,:,0] + image[:,:,1]+ image[:,:,2]
plt.imshow(grigi)
```

## Es2

colori invertiti

```python
invertiti = image.copy()
invertiti[:,:,0] = image[:,:,1] 
invertiti[:,:,1] = image[:,:,2] 
invertiti[:,:,2] = image[:,:,0] 
plt.imshow(invertiti)
```

## Es3

```python
scacchiera = image[1300:1400, 230:400].copy()
alt, larghezza = scacchiera.shape()
scacchiera[0::2, 1::2, 0] = 0
scacchiera[1::2, 0::2, 0] = 0
scacchiera[0::2, 1::2, 1] = 0
scacchiera[1::2, 0::2, 1] = 0
scacchiera[0::2, 1::2, 2] = 0
scacchiera[1::2, 0::2, 2] = 0
plt.imshow(scacchiera)
```
