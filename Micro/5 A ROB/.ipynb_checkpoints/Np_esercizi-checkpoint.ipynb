# La funzione `np.roll()`

### shifta le posizione dell'array

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.arange(1,6)
array
```

```python
np.roll(array, 1) 
```

### Dato un array di numeri, usare *numpy* per determinare se l'array è monotono crescente

```python
array2 = np.arange(1, 6)
array2
```

```python
if np.count_nonzero((np.roll(array2, -1) > array2)[:-1]) == (len(array) -1):
    print("l'array è monotono")
else:
    print("l'array non è monotono")
```

```python
array2d = np.random.randint(0, 3, (20,20))
plt.imshow(array2d)
```

```python
array2d_new = np.roll(array2d, 5, axis = 0)
plt.imshow(array2d_new)
```

## Caricare un'immagine rgb, rgba

il canale R rimane così
il canale G shifta di n/3 lungo le colonne 
il canale B shifta di 2* (n_colonne / 2)

```python
image = plt.imread("./lancia_delta.jpeg")
image_shift = image.copy()
```

```python
image_shift[:,:, 1] = np.roll(image[:,:,1], image.shape[1] // 3)
image_shift[:,:, 2] = np.roll(image[:,:,2], 2*(image.shape[1]//2))
plt.imshow(image_shift)
```

```python
array = np.array([10,2,1,9,3,6])
array_sx = np.roll(array, -1)
array_dx = np.roll(array, 1)
(array_sx < array)[:-1]
(array_dx < array)[1:]
array_and = np.logical_and((array_sx < array)[:-1], (array_dx < array)[:-1])
np.where(array_and)
```

### es 9

```python
dado = np.array([1,2,3,4,5,6])
trucA = [1/8,1/8,1/8,1/8,1/4,1/4]
tiriA = np.random.choice(dado, (10,1000), p=trucA)
tiriB = np.random.choice(dado, (10,1000))

vincite = tiriA > tiriB
pareggio = tiriA == tiriB
perso = tiriA < tiriB

punteggioA = np.count_nonzero(vincite) + 0.5 * np.count_nonzero(pareggio)
punteggioB = np.count_nonzero(perso) + 0.5 * np.count_nonzero(pareggio)


punteggioA,punteggioB, punteggioA + punteggioB
```

```python
print("vince A" if punteggioA > punteggioB else "Vince B" if punteggioA < punteggioB else "Pareggio")
```

# Es massimi locali della funzione $y = sin(x)$

```python
x = np.linspace(0, 4*np.pi, 400)
y = np.sin(x)
y_sx = np.roll(y, -1)
y_dx = np.roll(y, 1)
(y_sx < y)[:-1]
(y_dx < y)[1:]
y_and = np.logical_and((y_sx < y)[:-1], (y_dx < y)[:-1])
p_max = np.where(y_and)
```
