Il file `Netflix-Dataset-Latest-2021.csv` contiene svariati dati di film e serie Netflix.

1) Utilizza matplotlib creare un grafico a dispersione che mostri la correlazione tra `IMDb Score` e `Awards Received`. I punti del grafico devono avere un colore che dipenda dal numero di voti ricevuti `IMDb Votes`. I grafici devono avere titolo, griglia ed etichette sugli assi.

2) Traccia lo stesso grafico del punto 1 facendo in modo che la dimensione dei punti sia proporzionale al numero di voti `IMDb Votes`.

3) Traccia in un'unica figura due due istogrammi che rappresentino rispettivamente le distribuzioni di `IMDb Score` e `Awards Received`.

```python
dati = pd.read_csv("./Netflix-Dataset-Latest-2021.csv")

IMDb_Score = dati["IMDb Score"].values
Awards = dati["Awards Received"].values
Voti = dati["IMDb Votes"].values

dati.head(5)
```

## Punto 1

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(IMDb_Score, Awards, c = Voti, alpha = 0.7, zorder = 4)
fig.set_facecolor("Yellow")
ax.set_title("Correlazione tra IMDb Score e Premi ricevuti", fontsize = 18)
ax.set_xlabel("IMDb Score", fontsize = 14)
ax.set_ylabel("Premi ricevuti", fontsize = 14)
ax.grid(True, color = "k", alpha = 0.6, zorder = -4)
plt.show()
```

## Punto 2

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(dati["IMDb Score"], Awards, c = Voti, alpha = 0.7, s = Voti / 5000, zorder = 4)
fig.set_facecolor("Orange")
ax.set_title("Correlazione tra IMDb Score e Premi ricevuti", fontsize = 18)
ax.set_xlabel("IMDb Score", fontsize = 14)
ax.set_ylabel("Premi ricevuti", fontsize = 14)
ax.grid(True, color ="k", alpha = 0.6, zorder = -4)
plt.show()
```

## Punto 3

```python
fig, axs = plt.subplots(nrows= 2, ncols= 1, figsize=(9,8))

axs[0].hist(IMDb_Score, bins = len(np.unique(IMDb_Score)), label = "IMDb Score", edgecolor = "black", color = "red")
axs[1].hist(Awards, bins = len(np.unique(Awards)), label = "Awards Received", edgecolor = "black", color = "Blue")

axs[0].set_xlabel("Punteggio IMBd", fontsize = 14, color = "white")
axs[0].set_ylabel("Quantità di film", fontsize = 14, color = "white")

axs[1].set_xlabel("Numero di premi", fontsize = 14, color = "white")
axs[1].set_ylabel("Numero di film", fontsize = 14, color = "white")

axs[0].legend(loc = "upper right", fontsize = 12, edgecolor = "black")
axs[1].legend(loc = "upper right", fontsize = 12, edgecolor = "black")

fig.set_facecolor("Green")

plt.suptitle("Istogrammmi di Punteggio IMDb e Premi ricevuti dai film", fontsize = 18, y = 0.93, color = "white")
plt.show()
```
