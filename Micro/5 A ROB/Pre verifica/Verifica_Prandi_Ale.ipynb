## Es 1

```python
pokemon = pd.read_csv('./pokemon.csv')
pokemon.head(10)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.hist(pokemon["Type1"].values, bins=len(np.unique(pokemon["Type1"].values)), edgecolor = "black") 
ax.set_xticks(ax.get_xticks(), ax.get_xticklabels(), rotation=45, ha='right')
ax.set_title("Numero di pokemon per ogni tipo", fontsize = 24)
ax.set_ylabel("Numero di pokemon", fontsize = 18)
plt.show()
```
