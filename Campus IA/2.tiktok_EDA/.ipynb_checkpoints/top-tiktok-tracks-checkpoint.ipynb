```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('tiktok.csv')
```

```python
df
```

```python
df['track_name'].value_counts().nlargest(20)
```

```python
#top 20 tracks
plt.bar(df['track_name'][:20],df['track_name'].value_counts().nlargest(20))
plt.xticks(rotation=90)
plt.show()
```

```python
#top 50 artists
plt.bar(df['artist_name'][:50],df['track_name'].value_counts().nlargest(50))
plt.xticks(rotation=90)
plt.show()
```

```python
df.corr()
```
